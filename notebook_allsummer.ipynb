```python
import pandas as pd
import os

def save_dict_to_excel(dictionary, filename, folder_path):
    filepath = os.path.join(folder_path, filename)
    df = pd.DataFrame.from_dict(dictionary)
    df.to_excel(filepath)

# Esempio di utilizzo

folder_path = os.path.abspath(os.path.join(os.getcwd()))
save_dict_to_excel(isola.data, 'output.xlsx', folder_path)
```
